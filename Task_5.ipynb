```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('deliveries.csv')
df1 = pd.read_csv('matches.csv')
```

```python
df.batting_team.replace({'Rising Pune Supergiants':'Rising Pune Supergiant'}, inplace=True)
df.bowling_team.replace({'Rising Pune Supergiants':'Rising Pune Supergiant'}, inplace=True)
df1.team1.replace({'Rising Pune Supergiants':'Rising Pune Supergiant'}, inplace=True)
df1.team2.replace({'Rising Pune Supergiants':'Rising Pune Supergiant'}, inplace=True)
df1.toss_winner.replace({'Rising Pune Supergiants':'Rising Pune Supergiant'}, inplace=True)
df1.winner.replace({'Rising Pune Supergiants':'Rising Pune Supergiant'}, inplace=True)
df_player_dismissed=df.player_dismissed.dropna()
df_player_dismissed.drop_duplicates(inplace=True)
lst=list(df_player_dismissed)
df.player_dismissed.fillna(value=0, inplace=True)
df.player_dismissed.replace(to_replace=lst, value=1,inplace=True)
df1.drop(columns=['umpire1','umpire2', 'umpire3'],inplace=True)
```

# Bowler Dataframe

```python
df['Runs'] = df.total_runs-df.bye_runs-df.legbye_runs-df.penalty_runs
Overs = (df.bowler.value_counts())/6
wick = df.iloc[:,[8,18]]
w1 = wick.groupby('bowler').player_dismissed.sum()
w = pd.DataFrame(data=w1)
match = pd.DataFrame(data = df.groupby('bowler').match_id.nunique())
overs = pd.DataFrame(data=Overs).reset_index()
overs.rename(columns={'index': 'bowler', 'bowler':'Overs'}, inplace=True)
bowlers_data = df.groupby(['bowler'])['Runs'].sum()
bow = pd.DataFrame(data=bowlers_data, columns = ['Runs'])
ex = pd.DataFrame(data=df.groupby('bowler').extra_runs.sum())

d1 = pd.merge(bow, overs, on='bowler')
d2 = pd.merge(d1, w, on = 'bowler')
d3 = pd.merge(d2, match, on='bowler')
bowler = pd.merge(d3, ex, on = 'bowler')
bowler['Economy'] = bowler['Runs']/bowler['Overs']
bowler.rename(columns={'Runs':'Runs Conceded','bowler':'Bowler', 'player_dismissed': 'Wickets', 'match_id':'Matches', 'extra_runs': 'Extras'}, inplace=True)
bowler
```

# Batsman Dataframe

```python
runs = pd.DataFrame(data = df.groupby(['batsman']).batsman_runs.sum())
matches = pd.DataFrame(data = df.groupby('batsman').match_id.nunique())
d1 = pd.merge(runs, matches, on = 'batsman')
d1['average'] = d1['batsman_runs']/d1['match_id']
mom = pd.DataFrame(data = df1.player_of_match.value_counts()).reset_index()
mom.rename(columns = {'index':'batsman'}, inplace=True)
d2 = pd.merge(d1, mom, on='batsman')
dismissal = df.groupby('batsman').dismissal_kind.agg(pd.Series.mode).to_frame().reset_index()
d3 = pd.merge(d2, dismissal, on = 'batsman')
d3.rename(columns={'batsman': 'Batsman', 'batsman_runs':'Runs', 'match_id':'Matches Played', 'average':'Average', 'player_of_match':'Player of The Match', 'dismissal_kind':'Most Common Dismissal Kind'}, inplace=True)
d3
```
